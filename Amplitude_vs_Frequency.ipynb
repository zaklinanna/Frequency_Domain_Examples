```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Paths to the CSV files
csv_files = [
    'Amph_Step_Down_0_min1.csv',
    'Amph_Step_Down_1_min1.csv',
    'Amph_Step_Up_0_1.csv',
    'Amph_Step_Up_min1_1.csv'
]

# Labels for the plots
plot_labels = ['Dataset A', 'Dataset B', 'Dataset C', 'Dataset D']


# Create a 2x2 grid of plots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.flatten()

for i, file in enumerate(csv_files):
    # Read the CSV file
    data = pd.read_csv(file)

    # Extract frequency values
    frequency = data.iloc[:, 0]

    # Extract amplitude values for each T value
    amplitude_columns = data.columns[1:]
    amplitudes = data[amplitude_columns]

    # Plot the data on the corresponding subplot
    for column in amplitude_columns:
        axes[i].plot(frequency, amplitudes[column], label=column)

    # Add labels and legend to each subplot
    axes[i].set_xlabel('Frequency (d$^{-1}$)')
    axes[i].set_ylabel('Amplitude (C)')
    axes[i].legend()
    axes[i].grid(True)

    # Set x-axis limits to positive frequencies only
    axes[i].set_xlim(left=0.5, right=3)
    axes[i].set_ylim(top=4, bottom=0)
    axes[i].set_title(f'{plot_labels[i]}')

# Adjust layout
plt.tight_layout()

```
